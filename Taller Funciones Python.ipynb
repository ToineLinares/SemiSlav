```python
#1
def time_converter(time):
    
    hour = [time[0], time[1]]
    minute = [time[3], time[4]]

    for i in range(len(hour)):
        hour[i] = int(hour[i])
    for j in range(len(minute)):
        minute[j] = int(minute[j])


    hora = hour[0]*10 + hour[1]
    minuto = minute[0]*10 + minute[1]

    a = ' AM'
    if  hora == 0:
        hora = 12
        h = str(hora) +':'+ str(minute[0])+str(minute[1]) + a
    elif hora!= 0 and hora < 12:
        h = str(hora)+':'+ str(minute[0])+str(minute[1]) + a
    elif hora == 12:
        a = ' PM'
        h = str(hora)+':'+ str(minute[0])+str(minute[1]) + a
    else:
        hora = hora - 12
        a = ' PM'
        h = str(hora)+':'+ str(minute[0])+str(minute[1]) + a
    return h
```
